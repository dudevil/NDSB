```python
import numpy as np
import matplotlib.pyplot as plt

BEST_VAL_ERROR = 1.27
```

```python
result = np.load("../data/tidy/64_3convlayers.npy")

figure(figsize=(8,6))
test_err = plt.plot(result[0], result[1], 'r', label="Validation error")
train_err =  plt.plot(result[0], result[2], 'b', label="Train error")
plt.xlabel("Gradient step")
plt.ylabel("Negative loglikelyhood")
plt.title("Validation/train error 3 convlayers 64x64 images")
plt.text(19000, 3.6,"Min validation error: %f" % result[1].min())
plt.text(19000, 3.45, "Momentum: 0.9")
plt.text(19000, 3.3, "Learning rate: 0.004")
plt.axhline(BEST_VAL_ERROR, linestyle='--', color='g', label="Best-so-far")
plt.legend()
plt.grid()
plt.show()
```
